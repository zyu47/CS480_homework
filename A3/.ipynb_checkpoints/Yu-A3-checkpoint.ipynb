## Methods

Independent viable (sample) was referred as $\Xv$, and dependent viable (target) was referred as $\Tv$. 

**For Ridge Regression:**

$\Xv$ was first standardized using the following formula:
$$\boldsymbol{Z} = (\Xv-\bar{\Xv})/\sigmav$$
Standardized $X$ was appended with one column of ones to calculate bias. Linear regression was solved using the following formula:
$$\boldsymbol{\omega} = (\boldsymbol{Z}^T \boldsymbol{Z} + \lambda\boldsymbol{I})^{-1} \boldsymbol{Z}^T \Tv$$

**For Neural Network:**

We add one or more layers of non-linear functions, denoted as $\boldsymbol{Z}$, to the model where
$$\boldsymbol{Z} = tanh(\Xv\boldsymbol{V})$$
and we model the output as following
$$\boldsymbol{Y} = \boldsymbol{Z}\boldsymbol{W}$$

To solve $\boldsymbol{V}$ and $\boldsymbol{Z}$, we minimize the mean squared error between each target value $t_{n,k}$ and output (predicted) value $y_{n,k}$, i.e.
$$
\begin{align*}
  E &= \frac{1}{N} \frac{1}{K} \sum_{n=1}^N \sum_{k=1}^K ( t_{n,k} - y_{n,k})^2
\end{align*}
$$
by gradient descent using the following formula:
$$
\begin{align*}
  v_{j,m} &\leftarrow v_{j,m} - \rho_h \frac{\partial E}{\partial v_{j,m}}\\
  w_{m,k} &\leftarrow w_{m,k} - \rho_o \frac{\partial E}{\partial w_{m,k}}
\end{align*}
$$
where $\rho_h$ and $\rho_o$ are small gradient ratios.

By solving the equations, we get the updating policy for V and W:
$$
\begin{align*}
      v_{j,m} & \leftarrow v_{j,m} + \rho_h \frac{1}{N} \frac{1}{K} \sum_{n=1}^N \sum_{k=1}^K (t_{n,k} - y_{n,k})   w_{m,k} (1-z_{n,m}^2) \tilde{x}_{n,j}\\
 w_{m,k} & \leftarrow w_{m,k} + \rho_o \frac{1}{N}  \frac{1}{K}  \sum_{n=1}^N (t_{n,k} - y_{n,k}) \tilde{z}_{n,m}
    \end{align*}
$$

## Functions

```python
import neuralnetworks as nn
import imp
imp.reload(nn)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def trainLinear(X, T, parameters):
    X_mean = np.mean(X, axis = 0)
    X_std = np.std(X, axis = 0)
    stand_X = (X- X_mean) / X_std
    X_input = np.hstack((np.ones((stand_X.shape[0], 1)), stand_X))
    lambdaDiag = np.eye(X_input.shape[1]) * parameters
    lambdaDiag[0,0] = 0.0
    w = np.linalg.lstsq (np.dot(X_input.T, X_input) + lambdaDiag, np.dot(X_input.T, T))[0]
    return {'means': X_mean, 'stds': X_std, 'w': w}
```

```python
def evaluateLinear(model, X, T):
    X_mean = model['means']
    X_std = model['stds']
    stand_X = (X - X_mean) / X_std
    X_input = np.hstack((np.ones((stand_X.shape[0], 1)), stand_X))
    predict = np.dot(X_input, model['w'])
    return np.sqrt(np.mean((predict - T)**2))
```

```python
def trainNN(X, T, parameters):
    nnet = nn.NeuralNetwork(X.shape[1],parameters[0],T.shape[1])
    nnet.train(X, T, nIterations = parameters[1])
    return nnet
```

```python
def evaluateNN(model, X, T):
    predict = model.use(X, False)
    return np.sqrt(np.mean((predict - T)**2))
```

```python
def trainValidateTestKFolds(trainf,evaluatef,X,T,parameterSets,nFolds,
                            shuffle=False,verbose=False):
    # Randomly arrange row indices
    rowIndices = np.arange(X.shape[0])
    if shuffle:
        np.random.shuffle(rowIndices)
    # Calculate number of samples in each of the nFolds folds
    nSamples = X.shape[0]
    nEach = int(nSamples / nFolds)
    if nEach == 0:
        raise ValueError("partitionKFolds: Number of samples in each fold is 0.")
    # Calculate the starting and stopping row index for each fold.
    # Store in startsStops as list of (start,stop) pairs
    starts = np.arange(0,nEach*nFolds,nEach)
    stops = starts + nEach
    stops[-1] = nSamples
    startsStops = list(zip(starts,stops))
    # Repeat with testFold taking each single fold, one at a time
    results = []

    # COMPLETE THIS FUNCTION BY IMPLEMENTING THE FOLLOWING STEPS.
    
    for testfold in range(nFolds):
        best_par = None
        best_error = None
        for parmSet in parameterSets:  # For each set of parameter values, called parmSet
            sum_error = 0
            for valfold in range(nFolds):
                if valfold == testfold:
                    continue
                trainfold = np.setdiff1d(range(nFolds), [testfold, valfold])
                train_row = []
                for tf in trainfold:
                    a, b = startsStops[tf]
                    train_row += rowIndices[a:b].tolist()
                X_train = X[train_row, :]
                T_train = T[train_row, :]
                
                a, b = startsStops[valfold]
                val_row = rowIndices[a:b].tolist()
                X_val = X[val_row, :]
                T_val = T[val_row, :]
                
                model = trainf(X_train, T_train, parmSet)                
                sum_error += evaluatef(model, X_val, T_val)
                
            mean_error = sum_error / (nFolds - 1)
            if best_error == None or mean_error < best_error:
                best_error = mean_error
                best_par = parmSet
        
        trainfold_new = np.setdiff1d(range(nFolds), [testfold])
        train_row_new = []
        for tf in trainfold_new:
            a, b = startsStops[tf]
            train_row_new += rowIndices[a:b].tolist()
        X_train_new = X[train_row_new, :]
        T_train_new = T[train_row_new, :]
        
        a, b = startsStops[testfold]
        test_row = rowIndices[a:b].tolist()
        X_test = X[test_row, :]
        T_test = T[test_row, :]
        
        model_new = trainf(X_train_new, T_train_new, best_par)
        train_error = evaluatef(model_new, X_train_new, T_train_new)
        test_error = evaluatef(model_new, X_test, T_test)
        
        tmp = ((np.array([train_error, best_error, test_error])).flatten()).tolist()
        summary = [best_par] + tmp
        if verbose:
            print(summary)
        results.append(summary)
    return results
```

```python
### test linear regression
X = np.arange(100).reshape((-1, 1))
T = np.abs(X -50) + X

result = trainValidateTestKFolds(trainLinear, evaluateLinear ,X, T, range(0,101,10), nFolds=5, shuffle=False, verbose = True)
                                 
print('Linear Model\n{:^10s} {:>10s} {:>10s} {:>10s}'.format('lambda', 'train', 'validate', 'test RMSE'))
for x in result:
    print('{:^10.2f} {:>10.3f} {:10.3f} {:10.3f}'.format(*x))
```

```python
#Test neural network
import itertools
parms = list(itertools.product([2, 5, 10, 20, [5, 5], [10, 2, 10]], [10, 20, 100, 500]))
result = trainValidateTestKFolds(trainNN, evaluateNN, X, T, parms, nFolds=5, shuffle=False)
print('NN Model\n{:>30s} {:>10s} {:>10s} {:>10s}'.format('(Hidden Units, Iterations)', 'train', 'validate', 'test RMSE'))
for x in result:
    print('{:>30s} {:10.3f} {:10.3f} {:10.3f}'.format(str(x[0]), *x[1:]))
```

## Data

**Source:** http://archive.ics.uci.edu/ml/machine-learning-databases/parkinsons/telemonitoring/parkinsons_updrs.data

**Attributes:** The dataset was created by Athanasios Tsanas and Max Little of the University of Oxford, in collaboration with 10 medical centers in the US and Intel Corporation who developed the telemonitoring device to record the speech signals. The original study used a range of linear and nonlinear regression methods to predict the clinician's Parkinson's disease symptom score on the UPDRS scale.

This dataset is composed of a range of biomedical voice measurements from 42 people with early-stage Parkinson's disease recruited to a six-month trial of a telemonitoring device for remote symptom progression monitoring. There are 5875 instances (each representing one voice recording from those individuals), and 22 attributes. The main aim of the data is to predict the motor and total UPDRS scores ('motor_UPDRS' and 'total_UPDRS'), both of which are Parkinson's disease scale, from the 16 voice measures. The names and meanings of the 16 sample features are:

(1-5) *Jitter(%),Jitter(Abs),Jitter:RAP,Jitter:PPQ5,Jitter:DDP* - Several measures of variation in fundamental frequency.

(6-11) *Shimmer,Shimmer(dB),Shimmer:APQ3,Shimmer:APQ5,Shimmer:APQ11,Shimmer:DDA* - Several measures of variation in amplitude.

(12-13) *NHR,HNR* - Two measures of ratio of noise to tonal components in the voice.

(14) *RPDE* - A nonlinear dynamical complexity measure.

(15) *DFA* - Signal fractal scaling exponent.

(16) *PPE* - A nonlinear measure of fundamental frequency variation.

**Data plot:** 
See below for figures

**Description of Data:**
1. Most of the features do not possess an obvious positive or negative relationship with either output ('motor_UPDRS' and 'total_UPDRS'). However, two of the features: HNR and RPDE seem to be negatively and positively correlated with both outputs, respectively. 
2. Most of the data was not ubiquiously distributed across the feature range. In fact, most of them are localized in a lower range of the feature space. This may compromise the learning process, leading to higher probability of uncertainty in higher range of the feature space.
